## SMS Spam Detection – Data Cleaning

### Objective:
To preprocess the SMS spam dataset and prepare it 
for machine learning model training.

## Importing Library and Loading Dataset

```python
import pandas as pd
df = pd.read_csv("./spam.csv",encoding = "latin1")
df.head()
```

## Dataset Overview
Checking dataset structure and column information.

```python
df.shape
df.info()
```

## Missing Value Analysis
Identifying missing values in each column.

```python
df.isnull().sum()
```

## Removing Unnecessary Columns

Columns 'Unnamed: 2', 'Unnamed: 3', and 'Unnamed: 4' 

contain mostly null values and do not contribute 

to model training.

```python
df = df.drop(columns=["Unnamed: 2","Unnamed: 3","Unnamed: 4"] )
```

## Renaming Columns

Renamed columns for better readability.

v1 → Label  

v2 → Message

```python
df = df.rename(columns={"v1": "Label" , "v2": "Message"} )
df
```

## Duplicate Check

Identified and removed duplicate messages 

to avoid data redundancy.

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

## Label Encoding

Converted categorical labels into numeric format:

ham → 0
  
spam → 1

```python
df["Label"].value_counts()
```

```python
df['Label'] = df['Label'].map({'ham':0, "spam" : 1})
```

## Final Verification

Ensured no missing values and no duplicate records remain.

```python
df.isnull().sum()
df.duplicated().sum()
df.shape
```

## Exporting Cleaned Dataset

```python
df.to_csv("cleaned_SMS_spam.csv" , index = False)
```
