# SMS Text Dataset – Data Cleaning

Objective:

To clean and preprocess the SMS_Text dataset 

and prepare it for further analysis and model usage.

## Importing Library and Loading Dataset

```python
import pandas as pd
df = pd.read_csv("SMS_Text.csv")
df.head()
```

## Dataset Overview

Checking dataset shape and basic structure.

```python
df.shape
df.columns
```

## Missing Value Analysis

Verifying if any null values exist in the dataset.

```python
df.isnull().sum()
```

## Duplicate Check

Checked for duplicate messages and removed duplicates 

based on the 'data' column to ensure unique text entries.

```python
df.duplicated().sum()
df=df.drop_duplicates(subset="data")
```

## Renaming Column

Renamed 'data' column to 'message' 

for better clarity and consistency.

```python
df = df.rename(columns = {'data':"message"})
```

## Final Verification

Confirmed dataset integrity after cleaning.

```python
df.columns
df.shape
df.isnull().sum()
df.duplicated().sum()
```

## Exporting Cleaned Dataset

```python
df.to_csv("sms_text_cleaned.csv",index = False)
```
